```python
ret = []
for preference in preferences:
    data = requests.get(f'https://api.mercadopago.com/checkout/preferences/{preference["id"]}', headers={
        'Authorization': 'Bearer APP_USR-4846877984480703-062812-0851d62aca9c156ee183c30017844081-154849269'
    })
    fullPreference = data.json()
    items = []
    for item in fullPreference['items']:
        if item['id'] == '':
            continue
        itemId = json.loads(item['id'])
        options = []
        if (len(itemId["variantId"]) > 0):
            for variant in itemId["variantId"]:
                entry = client.entry(variant)
                options.append(entry.label)
        items.append({
            'item': item["title"], 
            'opcoes': options,
            'quantidade': item['quantity'],
            'price': float(item['unit_price'])
        })
    ret.append({
        'id': preference['id'],
        'name': fullPreference['payer']['name'],
        'email': fullPreference['payer']['email'],
        'total': sum([item['quantity'] * item['unit_price'] for item in fullPreference['items']]),
        'items': items
    })
    print(ret)
```
